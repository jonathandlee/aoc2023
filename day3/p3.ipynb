```python
import numpy as np
lines = [l.strip() for l in open("input.txt").readlines()]

nums = []
truenums = []
partlocs = []
mults = []
d = {}

for i in range(len(lines)):
  for j in range(len(lines[i])):
    if not lines[i][j].isnumeric() and lines[i][j] != '.':
      partlocs.append((i,j))
    if lines[i][j] == '*':
      mults.append((i,j))
for i in range(len(lines)):
  currnum = ""
  curlen = 0
  ispart = False
  ismult = False
  multcoords = []
  for j in range(len(lines[0])):
    if lines[i][j].isnumeric():
      # logic for if its a part
      for ii in range(i-1,i+2):
        for jj in range(j-1,j+2):
          if (ii,jj) in partlocs:
            ispart = True
          if (ii,jj) in mults and (ii,jj) not in multcoords:
            ismult = True
            multcoords.append((ii,jj))
      curlen += 1
      currnum += lines[i][j]
    else:
      if ispart:
        truenums.append(currnum)
      if curlen != 0:
        nums.append(currnum)
      if ismult:
        for ij in multcoords:
          if ij in d:
            d[ij].append(int(currnum))
          else:
            d[ij] = [int(currnum)]
      curlen = 0
      currnum = ""
      ispart = False
      #print(multcoords)
      ismult = False
      multcoords = []
  if curlen != 0:
    nums.append(currnum)
  if ispart:
    truenums.append(currnum)

  if ismult:
    for ij in multcoords:
      if ij in d:
        d[ij].append(int(currnum))
      else:
        d[ij] = [int(currnum)]
  curlen = 0
  currnum = ""
  
print(nums)
print(truenums)
print(partlocs)

print(d)

print(sum(int(i) for i in truenums))


acc = 0
for i in d:
  acc2 = 1
  for j in i:
    acc2 *= j
  acc += acc2
print(acc)
```
