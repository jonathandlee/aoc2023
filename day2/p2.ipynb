```python
lines = [l.strip() for l in open("input.txt").readlines()]
dic = {'red':12, 'blue':14, 'green':13}
counter = 0
pset = 0
for i in lines:
  minred = 0
  minblue= 0 
  mingreen = 0
  gameidl = i.index(' ')
  gameidr = i.index(':')
  gameid = (i[gameidl:gameidr])
  line = i[i.index(':')+2:]
  h = line.split(';')
  z = [hh.strip() for hh in h]
  won = False
  #print(z)
  for j in z:
    colors = j.split(',')
    colors = [color.strip() for color in colors]
    for jj in colors:
      mid = jj.index(' ')
      lside = jj[0:mid]
      rside = jj[mid+1:]
      if rside == 'green':
        mingreen = max(mingreen, int(lside))
      if rside == 'red':
        minred = max(minred, int(lside))
      if rside == 'blue':
        minblue = max(minblue, int(lside))
      if dic[rside] < int(lside):
        won = True
  if not won:
    counter += int(gameid)
  #print(minred,mingreen,minblue)
  pset += minred * mingreen * minblue
print(counter)
print(pset)


```
